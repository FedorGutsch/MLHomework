```python
from sklearn.cluster import KMeans
import seaborn as sns 
import pandas as pd
import numpy as np

df = pd.read_csv('iris.csv', delimiter=',')


df.head()
```

```python
X = df[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]

X
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
scaler = scaler.fit(X)
scaled_data = scaler.transform(X)


scaled_data
```

```python
kmeans = KMeans(n_clusters=3, random_state=1000)

model = kmeans.fit(X)
X['Label']=kmeans.labels_



sns.pairplot(data=X, hue='Label')

```

```python
sns.pairplot(data=df, hue='Class')
```

```python
X.loc[(X['Label']==1), 'Predicted']='Iris-virginica'
X.loc[(X['Label']==2), 'Predicted']='Iris-setosa'
X.loc[(X['Label']==0), 'Predicted']='Iris-versicolor'

X
```

```python
pd.crosstab(X['Predicted'],df['Class'])
```

```python
def predictClass(SL, SW, PL, PW):
    result = kmeans.predict([[SL, SW, PL, PW]])
    
    match result[0]:
        case 0:
            return 'Iris-versicolor'
        case 1:
            return 'Iris-virginica'
        case 2:
            return 'Iris-setosa'
        
print(predictClass(2,3,4,1))
```
