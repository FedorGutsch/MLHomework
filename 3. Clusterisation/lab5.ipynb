```python
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns


df=pd.read_csv('students.csv', delimiter=',')


# извлекаем данные по некоторым столбцам
df_cut=df[['Weight','Growth','Sex']]

# убираем записи с пустыми значениями
df_cut=df_cut.dropna()
df=df.dropna()

# делаем разделение на кластеры по параметрам (Вес, Рост)
kmeans = KMeans(n_clusters=2, random_state=1000)
kmeans.fit(df_cut[['Weight','Growth']])
df_cut['Label']=kmeans.labels_
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Label')
```

```python
df_cut.head(20)
```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Sex')
```

```python
df_cut.head()
```

```python
sum((df_cut['Sex']=='мужской')& (df_cut['Label']==0))
```

```python
sum((df_cut['Sex']=='женский') & (df_cut['Label']==1))
```

```python
# Отобразим матрицу ошибок:
man = [sum((df_cut['Sex']=='мужской')& (df_cut['Label']==1)),sum((df_cut['Sex']=='мужской')& (df_cut['Label']==0))]
woman = [sum((df_cut['Sex']=='женский')& (df_cut['Label']==1)),sum((df_cut['Sex']=='женский')&(df_cut['Label']==0))]

res_df = pd.DataFrame([['муж', *man],['жен', *woman],],columns = [ 'Class', 'Cluster(муж)','Cluster(жен)'])

```

```python
df_cut=df[['Weight','Shoe size','Sex']]

```

```python
df_cut.head()
print(df_cut.shape)

df_cut.dropna()
```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Shoe size', hue='Sex')
```

```python
from sklearn.preprocessing import StandardScaler

kmeans = KMeans(n_clusters=2, random_state=1)

X = df_cut[['Weight', 'Shoe size']]

scaler = StandardScaler()
scaler = scaler.fit(X)

scaled = scaler.transform(X=X)

model = kmeans.fit(scaled)
```

```python
df_cut['Label'] = model.labels_

sns.scatterplot(data=df_cut, x='Weight', y='Shoe size', hue='Label')
```

```python
man = [sum((df_cut['Sex']=='мужской')& (df_cut['Label']==1)),sum((df_cut['Sex']=='мужской')& (df_cut['Label']==0))]
woman = [sum((df_cut['Sex']=='женский')& (df_cut['Label']==1)),sum((df_cut['Sex']=='женский')&(df_cut['Label']==0))]

res_df = pd.DataFrame([['муж', *man],['жен', *woman],],columns = [ 'Class', 'Cluster(муж)','Cluster(жен)'])

res_df
```
