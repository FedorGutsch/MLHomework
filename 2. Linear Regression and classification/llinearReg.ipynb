```python
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LinearRegression

```

```python
df=pd.read_csv('students.csv', delimiter=',')

df_cut=df[['Growth','Shoe size']]
df_cut=df_cut.dropna()

sns.scatterplot(data=df_cut, x='Shoe size', y='Growth')
```

```python
linearRegression = LinearRegression()
results=linearRegression.fit(df_cut['Shoe size'].values.reshape(-1, 1), y=df_cut['Growth'].values)

df_cut['Shoe size'].values.reshape(-1,1)

results.coef_, results.intercept_
```

```python
df_cut['Predicted Growth']=results.predict(df_cut['Shoe size'].values.reshape(-1, 1))

# вычислим ошибку mean_absolute_error
from sklearn.metrics import mean_absolute_error
mean_absolute_error(df_cut['Growth'], df_cut['Predicted Growth'])
```

```python
df = pd.read_csv('students.csv', delimiter=',')
df_cut = df[['Age', 'Year of birth', 'Month of birthday']]

df.head(5)
```

```python
model = LinearRegression()

model.fit(X=df_cut[['Year of birth', 'Month of birthday']].values.reshape(-1, 2), y = df_cut[['Age']].values)
```

```python
model.intercept_, model.coef_
```

```python
print(f'{str(model.coef_[0][0])} * Year_Of_Birth + {str(model.coef_[0][1])} * Month_of_birthday + {str(model.intercept_[0])}' )
```

```python
df_test = pd.read_csv('students_test.csv', delimiter=',')

df_test_cut = df_test[['Age', 'Year of birth', 'Month of birthday']]


df_test_cut['Predicted'] = model.predict(df_test_cut[['Year of birth', 'Month of birthday']].values.reshape((-1, 2)))
```

```python
df_test_cut.head(40)
```

```python

mean_absolute_error(df_test_cut['Age'], df_test_cut['Predicted'])
```
