```python
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier

# подключение библиотеки с методом Ближайшего соседа
df=pd.read_csv('students.csv', delimiter=',')

df_cut=df[['Growth','Weight','Sex']]
df_cut=df_cut.dropna() # удаляем записи с пустыми значениями

```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Sex')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df_cut[['Weight', 'Growth']].values.reshape(-1, 2))

arr=scaler.transform(df_cut[['Weight', 'Growth']].values.reshape(-1, 2))
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(arr, y=df_cut['Sex'].values)
```

```python
df_test=pd.read_csv('students_test.csv', delimiter=',')
df_test_cut=df_test[['Growth','Weight','Sex']]
df_test_cut=df_test_cut.dropna()

# нормализация
arr_test=scaler.transform(df_test_cut[['Weight',
'Growth']].values.reshape(-1, 2))
# через функцию predict прогоняем объекты тестовой выборки

df_test_cut['Predicted']=model.predict(arr_test)
# Метрики качества задачи классификации

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
print(accuracy_score(df_test_cut['Predicted'],
df_test_cut['Sex']))
# процент правильно классифицированных объектов
print(precision_score(df_test_cut['Predicted'],
df_test_cut['Sex'],average=None, zero_division=1))
# tp / (tp + fp) отдельно по классам мужчин и женщин
print(recall_score(df_test_cut['Predicted'],
df_test_cut['Sex'],average=None, zero_division=1))
# tp / (tp + fn) отдельно по классам мужчин и женщин
# построим матрицу ошибок
pd.crosstab(df_test_cut['Predicted'],df_test_cut['Sex'])
```

```python

df_test_cut['Code']='0'
df_test_cut.loc[(df_test_cut['Sex']=='мужской') &
(df_test_cut['Predicted']=='женский'),'Code']='1'
df_test_cut.loc[(df_test_cut['Sex']=='женский') &
(df_test_cut['Predicted']=='мужской'),'Code']='2'

sns.scatterplot(data=df_test_cut, x='Weight', y='Growth',
hue='Code')
```
